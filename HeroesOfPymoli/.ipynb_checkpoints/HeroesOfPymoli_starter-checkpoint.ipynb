```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Count the number of unique Screen Names (SN) using nunique()
total_players = purchase_data["SN"].nunique()

# Store this in a DataFrame and print
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate summary info
num_items = purchase_data["Item Name"].nunique()
avg_price = purchase_data["Price"].mean()
num_purchases = purchase_data["Item Name"].count()
tot_revenue = purchase_data["Price"].sum()

# Store these in a DataFrame, format monetary values, and print
purchasing_analysis_total_df = pd.DataFrame({"Number of Unique Items": [num_items],
                                            "Average Price": [avg_price],
                                            "Number of Purchases": [num_purchases],
                                            "Total Revenue": [tot_revenue]})

purchasing_analysis_total_df["Average Price"] = purchasing_analysis_total_df["Average Price"].map("${:.2f}".format)
purchasing_analysis_total_df["Total Revenue"] = purchasing_analysis_total_df["Total Revenue"].map("${:.2f}".format)

purchasing_analysis_total_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Drop duplicate Screen Names (SN) so that each person's gender will only be counted once
genders_df = purchase_data.drop_duplicates(subset=["SN"])["Gender"]

# Create new data frame with only the counts of each Gender and rename column
gender_demographics_df = pd.DataFrame(genders_df.value_counts())
gender_demographics_df.rename(columns = {"Gender":"Total Count"}, inplace = True)

# Create new column for percentages and format as % with 2 decimal places
gender_demographics_df["Percentage of Players"] = gender_demographics_df["Total Count"] / total_players
gender_demographics_df["Percentage of Players"] = gender_demographics_df["Percentage of Players"].map("{:.2%}".format)
gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Group data by gender to perform calculations
gender_group = purchase_data.groupby(["Gender"])

# Calculate each column
purchase_counts = gender_group["Purchase ID"].count()
avg_purchase_price = gender_group["Price"].mean()
total_purchase_value = gender_group["Price"].sum()
# Note: for the following column, we divide by the number of unique individuals of each gender, 
# which is stored in gender_demographics_df["Total Count"]
avg_total_purchase_per_person = total_purchase_value / gender_demographics_df["Total Count"]

# Store these in a new DataFrame, format values, and print
purchasing_analysis_gender_df = pd.concat([purchase_counts, 
                                           avg_purchase_price, 
                                           total_purchase_value, 
                                           avg_total_purchase_per_person], axis=1, keys=["Purchase Count",
                                                                                         "Average Purchase Price", 
                                                                                         "Total Purchase Value", 
                                                                                         "Avg Total Purchase per Person"])

purchasing_analysis_gender_df["Average Purchase Price"] = purchasing_analysis_gender_df["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_gender_df["Total Purchase Value"] = purchasing_analysis_gender_df["Total Purchase Value"].map("${:.2f}".format)
purchasing_analysis_gender_df["Avg Total Purchase per Person"] = purchasing_analysis_gender_df["Avg Total Purchase per Person"].map("${:.2f}".format)

purchasing_analysis_gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create bins for ages (note: could use integers for some bins - 14, 19, 24, etc. since they are by default
# right inclusive, but this allows for any age that may be a float in the future)
bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 130]

# Create the names for the bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Create new DF without duplicate Screen Names (SN) so that each person's age will only be counted once
ages_df = purchase_data.drop_duplicates(subset=["SN"]).copy()

# Categorise players by Age Group
ages_df["Age Group"] = pd.cut(ages_df["Age"], bins, labels=group_names, include_lowest=True)
age_group = ages_df.groupby(["Age Group"])

# Calculate numbers by age group and add to a new DataFrame
age_count = age_group["Age"].count()
age_demographics_df = pd.DataFrame({"Total Count": age_count})
# and percentages
age_demographics_df["Percentage of Players"] = age_demographics_df["Total Count"] / total_players
age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].map("{:.2%}".format)
age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Group data by age groups to perform calculations
bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 130]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names, include_lowest=True)
age_groups = purchase_data.groupby(["Age Group"])

# Calculate each column
purchase_counts = age_groups["Purchase ID"].count()
avg_purchase_price = age_groups["Price"].mean()
total_purchase_value = age_groups["Price"].sum()
# Note: for the following column, we divide by the number of unique individuals of each age group, 
# which is stored in age_demographics_df["Total Count"]
avg_total_purchase_per_person = total_purchase_value / age_demographics_df["Total Count"]

# Store these in a new DataFrame, format values, and print
purchasing_analysis_age_df = pd.concat([purchase_counts, 
                                           avg_purchase_price, 
                                           total_purchase_value, 
                                           avg_total_purchase_per_person], axis=1, keys=["Purchase Count",
                                                                                         "Average Purchase Price", 
                                                                                         "Total Purchase Value", 
                                                                                         "Avg Total Purchase per Person"])

purchasing_analysis_age_df["Average Purchase Price"] = purchasing_analysis_age_df["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_age_df["Total Purchase Value"] = purchasing_analysis_age_df["Total Purchase Value"].map("${:.2f}".format)
purchasing_analysis_age_df["Avg Total Purchase per Person"] = purchasing_analysis_age_df["Avg Total Purchase per Person"].map("${:.2f}".format)

purchasing_analysis_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group data by Screen Name (SN) to perform calculations
sn_group = purchase_data.groupby(["SN"])

# Calculate each column
purchase_counts = sn_group["Purchase ID"].count()
avg_purchase_price = sn_group["Price"].mean()
total_purchase_value = sn_group["Price"].sum()

# Store these in a new DataFrame
top_spenders_df = pd.concat([purchase_counts,
                             avg_purchase_price,
                             total_purchase_value], axis=1, keys=["Purchase Count",
                                                                  "Average Purchase Price", 
                                                                  "Total Purchase Value"])
# Sort by Total Purchase Value, format values, and print
top_spenders_df.sort_values("Total Purchase Value", ascending=False, inplace=True)
top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:.2f}".format)

top_spenders_df.head(10)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group data by Item ID and Item Name to perform calculations, and extract Item Price column
price_group = purchase_data.groupby(["Item ID", "Item Name"])
# When extracting Item Price values, we take the mean, because some Items have different Prices for different purchases
item_price = price_group["Price"].mean()
purchase_count = price_group["Price"].count()
total_purchase_value = price_group["Price"].sum()

# Store these in a new DataFrame
most_pop_items_df = pd.concat([purchase_count, 
                               item_price, 
                               total_purchase_value], axis=1, keys=["Purchase Count", 
                                                                    "Item Price", 
                                                                    "Total Purchase Value"])

# Sort by Purchase Count, format values, and print
most_pop_items_df.sort_values("Purchase Count", ascending=False, inplace=True)
most_pop_items_df["Item Price"] = most_pop_items_df["Item Price"].map("${:.2f}".format)
most_pop_items_df["Total Purchase Value"] = most_pop_items_df["Total Purchase Value"].map("${:.2f}".format)

most_pop_items_df.head(10)
```
